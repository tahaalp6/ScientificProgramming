# Exercise 9.4

```python
import time
start = time.clock()
Z = np.random.normal(size=(n, n))

def normalize(x):
    fac = abs(x).max()
    x_n = x / x.max()
    return fac, x_n

x = np.array(1*n)

for i in range(8):
    x = np.dot(Z, x)
    lambda_1, x = normalize(x)
    
print('Largest Eigenvalue:', lambda_1)
print('Eigenvector:', x)
end = time.clock()
print("Current processor time (in seconds):", end - start)
```
